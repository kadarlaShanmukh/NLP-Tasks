```python
1. What is the difference between "Love" and "love" in NLP?
In NLP, "Love" and "love" are treated as two different words because computers are case-sensitive.
For us, both look similar, but for a machine:
"Love" (capital L) is one token
"love" (small l) is another token
This can increase the vocabulary size unnecessarily and may confuse the model.
So, in most cases, we convert everything to lowercase to make the text consistent and easier to process.


2. What happens if stopwords are not removed?
Stopwords are very common words like is, the, and, in, of, etc.
If we don’t remove them:
The text will contain a lot of unnecessary words
Important words may get less importance
The model may take more time to process
It can reduce the accuracy of the model

For example:
“I am going to the store”
After removing stopwords → “going store”
This keeps only the important information.


3. Two real-world scenarios where removing stopwords can be harmful
Even though stopwords are often removed, sometimes they are very important.
Scenario 1: Sentiment Analysis

Example:
“I am not happy”
If we remove stopwords → “happy”
This completely changes the meaning.
So, words like “not” are very important in sentiment analysis.


4. Difference between stemming and lemmatization
Both are used to convert words into their base form, but they work differently.

Stemming
Stemming simply cuts off the end of a word.
It is fast
But sometimes gives incorrect words

Example:
“running” → “run”
“studies” → “studi” (not a real word)


Lemmatization
Lemmatization is more advanced.
It uses a dictionary and grammar rules
It gives correct base words

Example:
“running” → “run”
“studies” → “study”
```

```python
import re

def preprocess_text(text):
    """
    This function cleans and preprocesses raw text data.
    It returns a list of cleaned tokens.
    """
    
    # 1. Handle empty input
    if not text or not isinstance(text, str):
        return []
    
    # 2. Convert to lowercase
    text = text.lower()
    
    # 3. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 4. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # 5. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 6. Handle repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 7. Remove special characters (keep only alphabets and spaces)
    text = re.sub(r'[^a-z\s]', '', text)
    
    # 8. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 9. Tokenization (split into words)
    tokens = text.split()
    
    # 10. Remove very short words (≤2), but keep 'no' and 'not'
    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ['no', 'not']:
            cleaned_tokens.append(word)
    
    return cleaned_tokens
```
